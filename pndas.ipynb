```python
virat_data["Pos"]=virat_data['Pos'].map({
    3.0:"Batting at 3",
    4.0:"Batting at 4",
    2.0:"Batting at 2",
    1.0:"Batting at 1",
    7.0:"Batting at 7",
    5.0:"Batting at 5",
    6.0:"Batting at 6"
})
print(virat_data[['Runs','Pos']])
```

```python
pos_counts=virat_data["Pos"].value_counts()
print(pos_counts)
print(type(pos_counts))
```

```python
pos_values=pos_counts.values
pos_index=pos_counts.index
fig=plt.figure(figsize=(10,7))
plt.pie(pos_values,labels=pos_index)
plt.show()
```

```python
def show_pie_plot(virat_data,key):
    count=virat_data[key].value_counts()
    count_labels=count.index
    count_values=count.values
    fig=plt.figure(figsize=(10,7))
    plt.pie(count_values,labels=count_labels)
    plt.show()
show_pie_plot(virat_data,"Opposition")
```

#  number of matches of ground

```python
run_at_pos=virat_data.groupby("Pos")['6s'].sum()
print(run_at_pos)
```

```python
run_at_pos_val=run_at_pos.values
run_at_pos_lab=run_at_pos.index
fig=plt.figure(figsize=(10,7))
plt.bar(run_at_pos_lab,run_at_pos_val,color="green",width=0.3)
plt.show()
```

```python
run_at_coun=virat_data.groupby("Opposition")['Runs'].sum()
print(run_at_coun)
```

```python
run_at_coun_val=run_at_coun.values
run_at_coun_lab=run_at_coun.index
fig=plt.figure(figsize=(10,7))
plt.bar(run_at_coun_lab,run_at_coun_val,color="red",width=0.3)
plt.show()
```
