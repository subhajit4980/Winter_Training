```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=np.arange(0,3*np.pi ,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
# print(x)
# print(y)
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.title("sine and cosine graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(["sine","cosine"])
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title("sine graph")
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title("cosine graph")
plt.show()
```

```python
y=(x**2)
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
def y_fun(x):
    return (-5*(x**2)+(6*x)+1)
y=y_fun(x)
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
def fun_y(x):
    return (1-(x**2)/2)
y=fun_y(x)
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x=np.linspace(-20,20,100)
def fun__Y(x):
    y=[]
    for i in x:
        r=i**2
        y.append(r)
    return y
y=fun__Y(x)
plt.plot(x,y)
plt.show()
```
